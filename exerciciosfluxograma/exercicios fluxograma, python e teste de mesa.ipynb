DESAFIOS - 3. QUANTO GASTARÁ EM GASOLINA


image.png

image.png


````python
#entrada
kp = float(input("Digite a quantidade  de quilometros percoridos"))
cg = float(input("Digite o consumo da gasolina em Km/l "))
pg = float(input("Digite o preço atual da gasolina "))

#processamento
GT = (kp/cg)*pg

#saida
print("Você gastará em média:R$  ",GT)
````

DESAFIOS - 4. CAUCULAR A VELOCIDADE MÉDIA DO CARRO

image.png

````python
# entrada
pp = float(input("Digite a quantidade de Km percorridos "))
hg = float(input("Digite apenas a quantidade de horas gastas nesse percurso "))
mg = float(input("Digite apenas a quantidade de minutos gastos "))

#processamento
TG = hg+(mg/60)
VM = pp/TG

#saida
print("A velocidade média do carro é de: ",VM)
````
