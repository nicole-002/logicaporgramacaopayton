3. PAGAMENTO DE COMISSÃO DE VENEDORES DE PEÇAS

DESAFIOS - 2.CAUCULA O VALOR QUE DEVE RECEBER PELAS HORAS EXTRAS

```python
#entrada
h = float(input("Digite a quantidade de horas trabalhadas "))
m = float(input("Digite a quantidade de minutos minutos trabalhados "))
gh = float(input("Digite quanto você ganha por hora "))

#processamento
M = m/60
D = h + M
G = (D*gh)
GT = G + G*70/100
#saida
print("Você deve ganhar:R$ ",GT)
print("por horas trabalhadas")
```

DESAFIOS - 3. QUANTO GASTARÁ EM GASOLINA

```python
#entrada
kp = float(input("Digite a quantidade  de quilometros percoridos na ida "))
cg = float(input("Digite o consumo da gasolina em Km/l "))
pg = float(input("Digite o preço atual da gasolina "))

#processamento
GG = kp/cg
GI = GG*pg
GT = GT*2

#saida
print("Você gastará em média:R$  ",GI)
print("na ida ou R$  ",GT)
print("para ida e volta")

```

DESAFIOS - 4. cAUCULAR A VELOCIDADE MEDIA DO CARRO

```python
# entrada
pp = float(input("Digite a quantidade de Km percorridos "))
hg = float(input("Digite apenas a quantidade de horas gastas nesse percurso "))
mg = float(input("Digite apenas a quantidade de minutos gastos "))

#processamento
TG = hg+(mg/60)
VM = pp/TG

#saida
print("A velocidade média do carro é de: ",VM)
```
