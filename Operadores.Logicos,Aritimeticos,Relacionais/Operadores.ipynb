#Operadores Aretiméticos

```python
#entrada
n1 = float(input("Digite um número"))
n2 = float(input("Digite um número"))

#processamento
soma= n1+n2
subtracao= n1-n2
multiplicacao= n1*n2
expoente= n1**n2
divisao= n1/n2
quociente= n1//n2
resto= n1%n2

#saida
print('Soma: ',soma,'\nSubtração: ',subtracao,
      '\nMultiplicacao: ',multiplicacao, '\nExpoente: ',expoente,
      '\nDivisao: ',divisao, '\nQuociente: ',quociente, '\nResto: ',resto)
```

#Operadores Relacionais

```python
#entrada
n1 = float(input("Digite um número A"))
n2 = float(input("Digite um número B"))

#processamento
igual= n1==n2
diferente= n1!=n2
maior= n1>n2
menor= n1<n2
maiorouigual= n1>=n2
menorouigual= n1<=n2

#saida
print("A é igual a B?: ",igual)
print("A é diferente de B?: ",diferente)
print("A é maior que B?: ",maior)
print("A é menor que B?: ", menor)
print("A é maior ou igual a B?: ",maiorouigual)
print("A é menor ou igual a B?: ", menorouigual)

```
